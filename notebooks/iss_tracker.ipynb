```python
import requests
import plotly.express as px
import datetime as dt
import os

print("✅ Imports OK")
print("Current folder:", os.getcwd())
print("Figures folder exists?", os.path.exists("../figures"))
```

```python
# Get live ISS position
response = requests.get("http://api.open-notify.org/iss-now.json")
data = response.json()

timestamp = dt.datetime.fromtimestamp(data["timestamp"])
lat = float(data["iss_position"]["latitude"])
lon = float(data["iss_position"]["longitude"])

print(f"🛰 ISS at {timestamp}: lat={lat}, lon={lon}")
```

```python
fig = px.scatter_geo(lat=[lat], lon=[lon], projection="natural earth",
                     title="Live ISS Position")
fig.update_traces(marker=dict(size=15, color="red"))
fig.show()

fig.write_image("../figures/iss_live.png", width=1200, height=600)
print("✅ Saved iss_live.png")
```

```python
# Get current ISS crew
response = requests.get("http://api.open-notify.org/astros.json")
crew = response.json()["people"]
print(f"👨‍🚀 {len([p for p in crew if p['craft'] == 'ISS'])} astronauts on ISS:")
for person in crew:
    if person["craft"] == "ISS":
        print(f"• {person['name']}")
```

```python
# Simple trail: ISS + nearby points
fig = px.scatter_geo(
    lat=[lat-0.1, lat, lat+0.1], 
    lon=[lon-0.1, lon, lon+0.1],
    projection="natural earth",
    title="ISS + Orbit Approximation"
)
fig.update_traces(marker=dict(size=[8,15,8], color="orange"))
fig.show()
fig.write_image("../figures/iss_orbit.png")
print("✅ Orbit saved")
```
